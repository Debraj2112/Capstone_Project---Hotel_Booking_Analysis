### Analysis 01: Booking Cancellation Analysis

```python
hotel_bookings_df
```

```python
resort_hotel = hotel_bookings_df.groupby('hotel').get_group('Resort Hotel')
resort_hotel
```

```python
city_hotel = hotel_bookings_df.groupby('hotel').get_group('City Hotel')
city_hotel
```

```python
resort_hotel_not_cancelled = resort_hotel[resort_hotel['is_canceled']==0].value_counts().sum()
resort_hotel_cancelled = resort_hotel[resort_hotel['is_canceled']==1].value_counts().sum()
print(resort_hotel_not_cancelled)
print(resort_hotel_cancelled)
```

```python
city_hotel_not_cancelled = city_hotel[city_hotel['is_canceled']==0].value_counts().sum()
city_hotel_cancelled = city_hotel[city_hotel['is_canceled']==1].value_counts().sum()
print(city_hotel_not_cancelled)
print(city_hotel_cancelled)
```

```python
hotel_cancellation_df = pd.DataFrame(columns = ['not_cancelled','cancelled','cancellation_%'], index = ['Resort_Hotel','City_Hotel'])

hotel_cancellation_df.loc['Resort_Hotel'] = [resort_hotel_not_cancelled,resort_hotel_cancelled, round((resort_hotel_cancelled/(resort_hotel_not_cancelled+resort_hotel_cancelled))*100,2)]

hotel_cancellation_df.loc['City_Hotel'] = [city_hotel_not_cancelled,city_hotel_cancelled,round((city_hotel_cancelled/(city_hotel_not_cancelled+city_hotel_cancelled))*100,2)]
```

```python
hotel_cancellation_df
```

```python
x1 = 100 - hotel_cancellation_df.loc['Resort_Hotel','cancellation_%']
y1 = 100 - hotel_cancellation_df.loc['City_Hotel','cancellation_%']

hotel_cancellation_df['confirmation_%'] = [x1,y1]
```

```python
hotel_cancellation_df
```

```python
#fig = plt.figure(figsize=(20, 7))
plt.pie(hotel_cancellation_df.loc['Resort_Hotel',:][2:],labels=list(hotel_cancellation_df.loc['Resort_Hotel',:][2:].index),colors=sns.color_palette('dark'), autopct='%.0f%%')
plt.title("Resort_Hotel_Cancellation")
plt.show()
```

```python
plt.pie(hotel_cancellation_df.loc['City_Hotel',:][2:],labels=list(hotel_cancellation_df.loc['City_Hotel',:][2:].index),colors=sns.color_palette('dark'), autopct='%.0f%%')
plt.title("City_Hotel_Cancellation")
plt.show()
```

```python
ax = sns.countplot(data = hotel_bookings_df, x = 'hotel', hue = 'is_canceled')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
hotel_cancellation_df
```

so from the above analysis we can conclude that city hotels are 12.28% more prone to cancellation compared to the resort hotels

### Analysis 02: Market Segment

```python
hotel_bookings_df.columns
```

```python
hotel_bookings_nc_df = hotel_bookings_df.groupby('is_canceled').get_group(0)

hotel_bookings_nc_df.head(20)
```

```python
market_segment_dict = {}

for item in list(set(hotel_bookings_nc_df['market_segment'])):
  market_segment_dict[item] = hotel_bookings_nc_df[hotel_bookings_nc_df['market_segment'] == item].value_counts().sum()

print(market_segment_dict)
```

```python
keys = [key for key in market_segment_dict]

print(keys)
```

```python
market_segment_df = pd.DataFrame(columns = ['no_of_bookings','market_share'],index=keys)
market_segment_df
```

```python
for item in keys:
  market_segment_df.loc[item] = [market_segment_dict[item],0]

market_segment_df
```

```python
def percentage(x,y):
  return round((x/y)*100,2)

total_no_of_bookings = sum(market_segment_df['no_of_bookings'])

for item in list(market_segment_df.index):
  market_segment_df.loc[item,'market_share'] = percentage(market_segment_df.loc[item,'no_of_bookings'],total_no_of_bookings)

market_segment_df
```

```python
market_segment_df_sorted = market_segment_df.sort_values(by='market_share',ascending=False)
market_segment_df_sorted
```

```python
sns.set(rc={'figure.figsize':(10,7.5)})
ax = sns.countplot(data = hotel_bookings_nc_df , x = 'market_segment' , order=hotel_bookings_nc_df['market_segment'].value_counts().index)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#fig = plt.figure(figsize=(20, 7))
plt.pie(market_segment_df_sorted['market_share'][0:5],labels=list(market_segment_df_sorted.index)[0:5],colors=sns.color_palette('dark'), autopct='%.0f%%')
plt.show()
```

```python
print(f"So we see that top 4 segments are having {round(sum(market_segment_df_sorted['market_share'][0:4]))} market share")
```

### Analysis 04: Weekday and Weekend Analysis

```python
stays_weekend_nights = sum(hotel_bookings_nc_df['stays_in_weekend_nights'])
stays_weekday_nights = sum(hotel_bookings_nc_df['stays_in_week_nights'])
```

```python
stays_df = pd.DataFrame(columns=['no_of_stays','percentage'],index=['stays_weekday_nights','stays_weekend_nights'])
```

```python
stays_df.loc['stays_weekend_nights'] = [stays_weekend_nights, percentage(stays_weekend_nights,(stays_weekend_nights+stays_weekday_nights))]
stays_df.loc['stays_weekday_nights'] = [stays_weekday_nights, percentage(stays_weekday_nights,(stays_weekend_nights+stays_weekday_nights))]
```

```python
stays_df
```

```python
#fig = plt.figure(figsize=(20, 7))
plt.pie(stays_df['percentage'],labels=list(stays_df.index),colors=sns.color_palette('dark'), autopct='%.0f%%')
plt.show()
```

```python
x = ['stays_weekend_nights','stays_weekday_nights']
y = [stays_weekend_nights,stays_weekday_nights]

sns.set(rc={'figure.figsize':(5,5)})
plt.bar(x,y)
plt.show()
```
